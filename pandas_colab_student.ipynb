### ‚úÖ Exercise 3
Drop rows where both `tip` & `size` are NaN; fill `size` by rounded mean per `day`.

```python
# YOUR WORK: Exercise 3
mask=tips_na['tip'].isna() & tips_na['size'].isna()
tips_cleaned=tips_na[~mask]
print(tips_cleaned)

tips_cleaned.loc[:,'size']=tips_cleaned['size'].fillna(
    tips_cleaned.groupby('day', observed=True)['size'].transform('mean').round()
)
```

### ‚úçÔ∏è Analysis (Missing Data) ‚Äî 3‚Äì5 sentences
- Compare `dropna` vs groupwise fill‚Äîwho shifts distribution more?
- What assumption does your fill strategy make?
- When would interpolation be inappropriate here?

1. Using dropna() removes rows with missing values, which can shift the distribution if NaNs are not randomly scattered.
2. Using groupwise fill (e.g., filling missing tip by median per day) preserves the overall shape better but assumes that tips are similar within the same day.
3. Interpolation would be inappropriate if the data are categorical or non-time-sequential (like day or size) because the method assumes smooth changes over index.

## 9) Time Series & Resampling (15m)

```python
tips_ts = tips.copy()
tips_ts['date'] = pd.to_datetime('2024-01-01') + pd.to_timedelta(np.arange(len(tips_ts)), unit='D')
tips_ts = tips_ts.set_index('date').sort_index()
weekly = tips_ts.resample('W')[['total_bill','tip']].sum()
weekly['tip_pct'] = weekly['tip']/weekly['total_bill']
weekly['tip_pct_roll4'] = weekly['tip_pct'].rolling(4, min_periods=1).mean()
weekly.head()
```

### ‚úçÔ∏è Analysis (Time Series & Resampling) ‚Äî 3‚Äì5 sentences
- What trend do you see weekly vs monthly?
- Interpret divergence between raw and rolling series.
- Why pick `W` vs `MS`?

1. Weekly (W) resampling smooths short-term fluctuations, while monthly (MS) resampling highlights broader trends.
2. Divergence between raw and rolling series shows local variability; spikes in raw data are smoothed in the rolling average.
3. W is chosen for more granularity and detecting weekly patterns, while MS aggregates data for strategic monthly insights.
4. Rolling windows reduce noise, making trends easier to visualize, but can lag peaks.

### üîé Quick checks
